```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

c_max = 10.0 #maximal allowed concentration of c
g_val = 1.0
k_a_val = 5.0
E_b = 5.0
kbs = np.linspace(0.01,20,30)
E_as = np.linspace(0.1,10,30)
res = []
nres = []

for k_b in kbs:
    for E_a in E_as:
        if(E_a/k_a_val>E_b/k_b) and (c_max*E_a/(c_max+k_a_val) >= g_val) and (c_max*E_b/(c_max+k_b) >= g_val):
            res.append((k_b,E_a,0))
        else:
            nres.append((k_b,E_a,0))

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
xs,ys,zs = zip(*res)
ax.scatter(xs,ys,zs)
ax.set_xlabel('$k_b$',size=20)
ax.set_ylabel('$E_a$',size=20)
```

```python
plt.scatter(xs,ys,c='blue',s = 15)
nxs,nys,nzs = zip(*nres)
plt.scatter(nxs,nys,c='red',s = 15)
plt.xlabel('$k_b$',size=15)
plt.ylabel('$E_a$',size=15)
plt.savefig('autotrophism.png')
```
